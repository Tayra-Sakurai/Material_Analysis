# Data Input

This outputs a 10-digit random numbers to `data.txt`.

```python
import random
import os
from pathlib import Path

if not Path('.', 'TestData').exists():
    os.mkdir('./TestData')

with open('./TestData/data.txt', 'w', encoding='utf-8', newline='\n') as file:
    for _ in range(10):
        file.write(f'{str(random.randrange(0, 10000000000)).zfill(10)}\n')
```
