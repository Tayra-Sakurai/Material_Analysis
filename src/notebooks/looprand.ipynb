# Loop Test

This outputs 10 random numbers to `test-(n).txt` (where `n` is in a range of 1 to 5) with separated by comma.

```python
import numpy as np
from pathlib import Path

rng = np.random.default_rng()
rands = rng.integers(0, 100, (5, 10,))

if not Path('.', 'TestData').exists():
    Path('.', 'TestData').mkdir()

for i, r in enumerate(rands):
    with open(f'./TestData/test-{i+1}.txt', 'w', encoding='utf-8', newline='\n') as file:
        file.writelines(
            [
                f'{num}\n' for num in r
            ]
        )
```
