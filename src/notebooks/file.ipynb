## File Operation

With using `open` built-in function, you can get access to the file you designate.

You should be careful of the encoding of the file. Especially, for Windows, the default encoding is UTF-8 with BOM (Byte Order Mark) and is not standard encoding.

The following example opens your designated file in encoding of UTF-8 **without BOM**.

```python
import pandas

# Ask user a file name.
fname = 'data.csv'

with open(fname, encoding='utf-8') as file:
    print(file.read())

print()

# Load the file as a DataFrame
df = pandas.read_csv(
    fname,
    encoding='utf-8',
    header=0
)
print(df)
```
