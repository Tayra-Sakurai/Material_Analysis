# The Conditional Statements

The conditional statements are very important for programming. The `if` statements processes the suite depending on the conditions.

An `if` statement is formed with the condition and suite. If the condition given to the `if` statement block is fulfilled, the suite is done.

When the conditions are not fulfilled, you can add an `else` statement to execute the processes.

Let's see how `if` statement works.

```python
# Define a variable a as 1
a = 1
# Define another variable as 3
b = 2

if a == b:
    # If a equals to b, this code block is executed.
    print("a equals to b.")
else:
    # Otherwise, this code block is executed.
    print("a does not equals to b.")
```

## Basic Conditional Operators

To use `if` statement, you should know conditional operators to specify the conditions.

#. The `==` operator checks if the value of the left and right hands of the operator equal.
#. The `>` and `<` operators compares the values and tells the value in the left hand is greater than or less than the right hand value, respectively.

The following example uses `elif` statement. This is executed if the previous conditions in the `if` and previous `elif` statements are not fulfilled but the condition given to the `elif` is fulfilled. This comes from the abbration of `else if`.

```python
# Let a be greater than b.
a = 10
b = 1

# Expected output: "a is greater than b."
if a == b:
    # Suites executed if a = b
    print("a equals to b.")
elif a < b:
    # Suites executed if a < b
    print("a is less than b.")
elif a > b:
    # Suites executed when a > b
    print("a is greater than b.")
```
