## Pseudorandom Numbers

[Python `random` module](https://docs.python.org/3.14/library/random.html) supports psedorandom number generators.

This program returns a random integer `n` which fulfils the following relation.
$$
0 \leq n < 100
$$

```python
import random
random.randrange(0, 100)
```
